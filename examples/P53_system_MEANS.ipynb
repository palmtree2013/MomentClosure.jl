```python
import means
import means.examples
import numpy as np
import matplotlib.pyplot as plt

sys = means.examples.MODEL_P53

# Values for parameters
ps = [90, 0.002, 1.7, 1.1, 0.93, 0.96, 0.01]
# Values for initial conditions
ic_values = [70, 30, 60]
# Time interval to solve on
tspan = np.arange(0, 200, 0.1)
```

```python
plt.figure(figsize=(8,4))

for closure in ["normal", "log-normal", "gamma"]:
    odes = means.mea_approximation(sys, 3, closure=closure, multivariate=True)
    simulation = means.Simulation(odes)
    trajectories = simulation.simulate_system(ps, ic_values, tspan)
    trajectories[0].plot(label=closure)

plt.margins(x=0)
plt.xlabel('Time [h]', fontsize=12)
plt.ylabel('Mean p53 molecule number', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.legend(fontsize=12)
plt.savefig("p53-Mdm2_3rd_order_expansion_MEANS.svg", bbox_inches='tight')
#plt.close()
plt.show()
```

```python
plt.figure(figsize=(8,4))

for closure in ["log-normal", "gamma"]:
    odes = means.mea_approximation(sys, 5, closure=closure, multivariate=True)
    simulation = means.Simulation(odes)
    trajectories = simulation.simulate_system(ps, ic_values, tspan)
    trajectories[0].plot(label=closure)

plt.margins(x=0)
plt.xlabel('Time [h]', fontsize=12)
plt.ylabel('Mean p53 molecule number', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.legend(fontsize=12)
plt.savefig("p53-Mdm2_5th_order_expansion_MEANS.svg", bbox_inches='tight')
#plt.close()
plt.show()
```
